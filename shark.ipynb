```python
#Function that remove make columns name lowercase
def update_columns_to_lower(shark):
    shark.columns = shark.columns.str.lower().str.strip()
    return shark

#Function that transform float into integer
def float_to_int(shark,column):
    shark[column] = shark[column].round(0).astype(int)
    return shark

#Function to remove spaces and make lowercase
def update_lowercase_removespace(shark):
    for column in shark.columns:
        if shark[column].dtype == object:
            shark[column] = shark[column].str.lower().str.strip()
    return shark

#Funcion that return the percentage of the whole column
def return_percentage(column,shark):
    percentages = shark[column].value_counts(normalize=True) * 100
    percentages = percentages.round(0).astype(int)
    return percentages

#Function that allow check the top 10 values for a respective column
def return_top_10(column,shark):
    top_10 = shark[column].value_counts(normalize=True).head(10) * 100
    top_10 = top_10.round(0).astype(int)
    return top_10

#Function that identify low occurence and modify to others when is lower than 1%
def group_low_occurences(column, shark):
    percentages = return_percentage(column, shark)
    low_occurrence_values = percentages[percentages < 1].index.tolist()
    shark_group = shark.copy()
    shark_group[column] = shark_group[column].apply(lambda x: "other" if x in low_occurrence_values else x)
    return shark_group

#Function that drop rows based in the number of ocurrences
def drop_rows_by_occurrences(column, shark, min_occurrences):
    value_counts = shark[column].value_counts()
    values_to_keep = value_counts[value_counts >= min_occurrences].index.tolist()
    shark_filtered = shark[shark[column].isin(values_to_keep)]
    return shark_filtered

#Function to identify value and categorize it
def categorize_activity(shark, column_name):
    fishing_related = ['fishing', 'spearfishing', 'kayak fishing', 'shark fishing', 'surf fishing', 'fishing for sharks',
                       'wade fishing', 'fishing for mackerel','spearfishing / free diving']

    water_sports_excluding_surf_swim = ['wading', 'bathing', 'diving', 'snorkeling', 'standing', 'scuba diving', 'body boarding',
                                        'boogie boarding', 'body surfing', 'kayaking', 'free diving', 'treading water',
                                        'fell overboard', 'pearl diving', 'surf skiing', 'windsurfing', 'floating', 'walking',
                                        'canoeing', 'rowing', 'sea disaster', 'playing', 'surf-skiing', 'kite surfing',
                                        'paddle boarding', 'freediving', 'sponge diving', 'diving for trochus', 'freedom swimming',
                                        'stand-up paddleboarding', 'sailing', 'fell into the water', 'sitting on surfboard',
                                        'diving for abalone', 'surfing (sitting on his board)', 'free diving for abalone',
                                        'floating on his back', 'skindiving', 'spearfishing on scuba', 'jumped overboard', 'shipwreck',
                                        'splashing', 'skin diving', 'boat']
    def activity_category(activity):
        if isinstance(activity, str):
            activity = activity.lower().strip()
            if activity in fishing_related:
                return 'fishing'
            elif activity in water_sports_excluding_surf_swim:
                return 'water_sports'
            elif activity == 'surfing':
                return 'surfing'
            elif activity == 'swimming':
                return 'swimming'
            else:
                return "uncategorized"
        else:
            return "uncategorized"
    shark['activity_group'] = shark[column_name].apply(activity_category)
    shark_filtered = shark[shark['activity_group'] != "uncategorized"]
    return shark_filtered

#Function to remove the invalid years
def remove_invalid_years(shark, column_name='year', min_valid_year=1800):
    shark[column_name] = pd.to_numeric(shark[column_name], errors='coerce')
    shark = shark.dropna(subset=[column_name])
    shark = shark[shark[column_name] >= min_valid_year]
    return shark

#Function to remove the invalid years
def remove_invalid_yesno(shark, column_name='yes/no'):
    test_change = shark[column_name].isin(['y', 'n'])
    return shark[test_change]

#Function to convert country to continent
test_list = []

def country_to_continent(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name, cn_name_format="default")
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name.lower()
    except:
        test_list.append(country_name)
        return "unknown"

#Fix countries
def correct_country_name(country_name):
    if isinstance(country_name, str):
        country_name_lower = country_name.strip().lower()
        if country_name_lower in country_corrections:
            return country_corrections[country_name_lower]
        return country_name
    else:
        return "Others"

#Separate Month from Date
def separate_month(date):
    try:
        month = date.split('-')[1]
        if len(month) == 3:
            if month == 'jan':
                month = "01"
            elif month == 'feb':
                month = "02"
            elif month == 'mar':
                month = "03"
            elif month == 'apr':
                month = "04"
            elif month == 'may':
                month = "05"
            elif month == 'jun':
                month = "06"
            elif month == 'jul':
                month = "07"
            elif month == 'aug':
                month = "08"
            elif month == 'sep':
                month = "09"
            elif month == 'oct':
                month = "10"
            elif month == 'nov':
                month = "11"
            elif month == 'dec':
                month = "12"
            return month
        else:
            return None
    except:
        return None

#Function to calculate risk
def calculate_adjusted_risk_score(row):
    weight_y = 2  # Weight for 'yes' incidents
    weight_n = 1  # Weight for 'no' incidents
    # Adjust the risk score calculation
    raw_score = (row['total_y'] * weight_y) + (row['total_n'] * weight_n)
    return raw_score / row['total_incidents']  # Divide by the total number of incidents
```

```python
#Remove space and make content lowercase
shark = update_lowercase_removespace(shark)

#Make lowercase the columns
shark = update_columns_to_lower(shark)

#Remove the invalid years
shark = remove_invalid_years(shark, 'year', 1800)

#Turn the year in integer for better management
shark = float_to_int(shark,'year')

#Create a activity_group for categorize the activities
shark = categorize_activity(shark, 'activity')

#Fix the countries
shark['country'] = shark['country'].apply(correct_country_name)

#Create a continent for better organization of the countries
shark['continent'] = shark['country'].apply(country_to_continent)

#Drop countries that could not be recognized
shark = shark[shark['continent'] != "others"]
shark = shark[shark['continent'] != "unknown"]

#Create a new index for future management
shark = shark.reset_index(drop=True)

#Set a month column
shark['month'] = shark['date'].apply(separate_month)

#Removing invalid data
shark = remove_invalid_yesno(shark, 'yes/no')

#Creating month-year
shark['month_year'] = shark.apply(lambda row: f"{row['year']}-{row['month']}", axis=1)

#Adjust the date
shark.dropna(subset=['month'], inplace=True)
shark['month_year'] = pd.to_datetime(shark['month_year'], format='%Y-%m')

#Deleting unnecessary columns
colums_to_drop = ['date','time','sex','age','type','state','location','name','injury','species','source']
shark = shark.drop(columns=colums_to_drop)

#Collect Incidents
incidents_per_country = shark.groupby('country')['country'].count().reset_index(name='incidents')
```

```python
shark['yes'] = (shark['yes/no'] == 'y').astype(int)
shark['no'] = (shark['yes/no'] == 'n').astype(int)

total_incidents_per_year = shark.groupby('year')['yes/no'].count().reset_index(name='total_incidents_year')
total_incidents_per_country = shark.groupby('country')['yes/no'].count().reset_index(name='total_incidents_country')
total_incidents_per_continent = shark.groupby('continent')['yes/no'].count().reset_index(name='total_incidents_continent')
total_incidents_per_activity_group = shark.groupby('activity_group')['yes/no'].count().reset_index(name='total_incidents_per_activity_group')

aggregated_shark_year = shark.groupby(['activity_group', 'year']).agg(
    total_y=('yes', 'sum'),
    total_n=('no', 'sum'),
    total_incidents=('yes', 'size')
).reset_index()

aggregated_shark_year = pd.merge(aggregated_shark_year, total_incidents_per_year, on='year', how='left')

aggregated_shark_year['rank'] = ((aggregated_shark_year['total_y'] * 2 + aggregated_shark_year['total_n']) / aggregated_shark_year['total_incidents_year']) * 100
aggregated_shark_year['rank'] = aggregated_shark_year['rank'].round(0).astype(int)

aggregated_shark_year['rank'] = aggregated_shark_year.groupby(['year'])['rank'].rank(method='dense', ascending=False).astype(int)

aggregated_shark_year.head()
```

```python
aggregated_shark_country = shark.groupby(['activity_group', 'country']).agg(
    total_y=('yes', 'sum'),
    total_n=('no', 'sum'),
    total_incidents=('yes', 'size')
).reset_index()

aggregated_shark_country = pd.merge(aggregated_shark_country, total_incidents_per_country, on='country', how='left')

aggregated_shark_country['rank'] = ((aggregated_shark_country['total_y'] * 2 + aggregated_shark_country['total_n']) / aggregated_shark_country['total_incidents_country']) * 100
aggregated_shark_country['rank'] = aggregated_shark_country['rank'].round(0).astype(int)

aggregated_shark_country['rank'] = aggregated_shark_country.groupby(['country'])['rank'].rank(method='dense', ascending=False).astype(int)

filtered_shark = aggregated_shark_country[aggregated_shark_country['activity_group'] == 'surfing']
top_rows = filtered_shark.nlargest(5, 'total_incidents')
display(top_rows)

filtered_shark = aggregated_shark_country[aggregated_shark_country['activity_group'] == 'swimming']
top_rows = filtered_shark.nlargest(5, 'total_incidents')
display(top_rows)

filtered_shark = aggregated_shark_country[aggregated_shark_country['activity_group'] == 'fishing']
top_rows = filtered_shark.nlargest(5, 'total_incidents')
display(top_rows)

filtered_shark = aggregated_shark_country[aggregated_shark_country['activity_group'] == 'water_sports']
top_rows = filtered_shark.nlargest(5, 'total_incidents')
display(top_rows)
```
